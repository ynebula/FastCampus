## 얕은 신경망을 이용한 다중 분류 문제

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 함수 구현
### Sigmoid 함수
$sigmoid(x) = 1/(1+e^{-x})$

```python
# x가 vertor로 가정
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### Softmax 함수
$softmax(x)_i = e^{x_i}/\sum{e^{x_j}}$

```python
def softmax(x):
    e_x = np.exp(x)
    return e_x / np.sum(e_x)
```

## 네트워크 구조 정의

```python
class ShallowNN:
    def __init__(self, num_input, num_hidden, num_output):
        # (M*N)* (N*1) => Input * Hidden_weight
        self.W_h = np.zeros((num_hidden, num_input), dtype=np.float32)
        self.b_h = np.zeros((num_hidden), dtype=np.float32)
        # output layer
        self.W_o = np.zeros((num_output, num_hidden), dtype=np.float32)
        self.b_o = np.zeros((num_output), dtype=np.float32)
    
    def __call__(self, x):
        h = sigmoid(np.matmul(self.W_h, x) + self.b_h)
        return softmax(np.matmul(self.W_o, h) + self.b_o)
```

## 데이터셋 가져오기, 정리하기

```python
# Import and organize dataset
dataset = np.load('ch2_dataset.npz')
inputs = dataset['inputs']
labels = dataset['labels'] # 10의 길이인 one-hot vector
```

## 모델 만들기

```python
model = ShallowNN(2, 128, 10)
```

## 사전에 학습된 파라미터 불러오기

```python
weights = np.load('ch2_parameters.npz')
model.W_h = weights['W_h']
model.b_h = weights['b_h']
model.W_o = weights['W_o']
model.b_o = weights['b_o']
```

## 모델 구동 및 결과 프린터

```python
outputs = list()
for pt, label in zip(inputs, labels):
    output = model(pt) # output은 softmax 결과로 나옴
    outputs.append(np.argmax(output)) # 처음이 가장 크며 0이 들어감
    print(np.argmax(output), label) # prediction과 label을 비교
    
outputs = np.stack(outputs, axis=0)
```

## 정답 클래스 스캐터 플랏

```python
plt.figure()
for idx in range(10): # 10개의 class가 있으므로
    mask = labels == idx
    plt.scatter(inputs[mask, 0], inputs[mask, 1]) # scatter가 돌면서 다른 색으로 표현됨
    
plt.title('true_label')
plt.show()
```

## 모델 출력 클래스 스캐터 플랏

```python
plt.figure()
for idx in range(10): # 10개의 class가 있으므로
    mask = outputs == idx
    plt.scatter(inputs[mask, 0], inputs[mask, 1]) # scatter가 돌면서 다른 색으로 표현됨
    
plt.title('model_outpus')
plt.show()
```
